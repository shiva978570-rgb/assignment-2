Check if the number is zero

Check if the number is negative

If the number is positive

```python
def check_number(num):
    if num == 0:
        return "Zero"
    elif num < 0:
        return "Negative"
    else:  # num > 0
        if num % 2 == 0:
            return "Positive Even"
        else:
            return "Positive Odd"
```

if a >= b and a >= c:

```python
def find_largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
```

The function uses nested if statements, meaning conditions are placed inside other conditions.

Step 1: Check if the student passed
if marks >= 50:


If marks are below 50, the student automatically gets "Fail".

If marks are 50 or more, we continue checking for higher grades.

Step 2: Check for higher grade ranges

Inside the marks >= 50 block:

If marks are 75 or more, we check again:

if marks >= 90:


90 or more → "A"

75–89 → "B"

If marks are 50–74, we return "C".

 Why This Approach Is Correct

The structure checks from lowest passing condition to highest grade.

Each nested condition further refines the result.

The ranges do not overlap.

Every possible mark falls into exactly one category.

The else statements ensure a grade is always returned.

```python
def calculate_grade(marks):
    if marks >= 50:
        if marks >= 75:
            if marks >= 90:
                return "A"
            else:
                return "B"
        else:
            return "C"
    else:
        return "Fail"
```

Input Validation (Negative Age Check)
1.  if age < 0:
    raise ValueError("Age cannot be negative.")


Age cannot logically be negative.

If a negative value is entered, we raise an error using ValueError.

This immediately stops execution and prevents invalid data from being processed.

2. Check if Minor
elif age < 18:
    return "Minor"


If age is between 0 and 17, the function returns "Minor".

3. Otherwise Adult
else:
    return "Adult"


If age is 18 or above, the function returns "Adult".

How Conditional Statements Help in Input Validation

Early validation: The first if statement checks for invalid input before any further processing.

Prevents incorrect results: Without validation, a negative age might incorrectly return "Minor".

Ensures logical correctness: Conditions allow us to enforce real-world rules (e.g., age cannot be negative).

Improves program reliability: Raising an error alerts the user immediately instead of silently accepting wrong data.

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    elif age < 18:
        return "Minor"
    else:
        return "Adult"
```

```python
print(check_age(15))   
print(check_age(21)) 
print(check_age(2))
```

Explanation of the Logic

This function uses both a loop and conditional statements.

1. Loop Through the List
for num in numbers:


The for loop processes each number in the list one by one.

2. Stop if Zero Appears
if num == 0:
    break


If a 0 is found, the break statement immediately stops the loop.

No further elements are processed after zero.

3. Ignore Negative Numbers
elif num < 0:
    continue


If the number is negative, continue skips the current iteration.

The loop moves to the next number without counting it.

4. Count Even Numbers
elif num % 2 == 0:
    count += 1


If the number is positive and divisible by 2, it is even.

We increase the counter by 1.

 Why This Works Correctly

The loop ensures every number is checked sequentially.

break guarantees processing stops at zero.

continue prevents negative numbers from affecting the count.

The conditional structure ensures each number is handled appropriately.

The function always returns the correct count of positive even numbers before zero appears.

```python
def count_even_numbers(numbers):
    count = 0
    
    for num in numbers:
        if num == 0:
            break          # Stop processing if zero appears
        elif num < 0:
            continue       # Ignore negative numbers
        elif num % 2 == 0:
            count += 1     # Count positive even numbers
    
    return count
```

```python
nums = [4, -2, 6, 3, 0, 8, 10]
print(count_even_numbers(nums))
```
